```python
# consider a different sampling method

def naive_rand_partition(n):
    """
    Returns a random partition of n
    """
    x = np.random.rand(n)
    x = x/ np.sum(x)
    return x
```

```python
simplex_samps_naive = np.concatenate([naive_rand_partition(2).reshape(1,-1) for _ in range(500000)])
```

```python
diff = simplex_samps_naive[:,1] - simplex_samps_naive[:,0]
fig, ax = plt.subplots()
ax.hist(diff, bins=100, density=True, alpha=0.6, color='g', label='diff')
plt.show()
```
